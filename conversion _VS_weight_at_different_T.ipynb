```
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# -------------------------------------------------------
# FIXED CONSTANTS
# -------------------------------------------------------
k0 = 10**2.845
Ea = 111e3
R = 8.314
P_total = 30 * 1e5  # Pa

# Reaction orders
n = 1.0
m = 0.0
l = -0.36
p = -0.09

# Inlet molar flowrates (mol/s)
F_CO0   = 1515/3600
F_H2O0  = 2222/3600
F_H2_0  = 5555/3600
F_CO2_0 = 505/3600
F_CH4_0 = 303/3600  # inert

# -------------------------------------------------------
# FUNCTIONS
# -------------------------------------------------------
def Kp(T):
    return np.exp(4577.8/T - 4.33)

def rate_constant(T):
    return k0 * np.exp(-Ea/(R*T))

# PFR model
def pfr_odes(W, y, T):
    F_CO, F_H2O, F_H2, F_CO2 = y
    Ftot = F_CO + F_H2O + F_H2 + F_CO2 + F_CH4_0

    y_CO  = F_CO  / Ftot
    y_H2O = F_H2O / Ftot
    y_H2  = F_H2  / Ftot
    y_CO2 = F_CO2 / Ftot

    P_CO  = y_CO  * P_total
    P_H2O = y_H2O * P_total
    P_H2  = y_H2  * P_total
    P_CO2 = y_CO2 * P_total

    beta = (P_CO2 * P_H2) / (P_CO * P_H2O * Kp(T))
    k = rate_constant(T)

    r_CO = k * (P_CO**n) * (P_H2O**m) * (P_CO2**l) * (P_H2**p) * (1 - beta)

    return [
        -r_CO,
        -r_CO,
        +r_CO,
        +r_CO,
    ]


# -------------------------------------------------------
# TEMPERATURE SWEEP (310°C to 420°C)
# -------------------------------------------------------
T_list_C = [200, 300, 310, 320, 340, 350]
colors = ['b', 'g', 'r', 'c', 'm', 'k']

plt.figure(figsize=(9,6))

for T_C, c in zip(T_list_C, colors):

    T = T_C + 273.15  # convert to K

    y0 = [F_CO0, F_H2O0, F_H2_0, F_CO2_0]

    # solve PFR
    sol = solve_ivp(
        lambda W, y: pfr_odes(W, y, T),
        [0, 800000],       # search range
        y0,
        max_step=20,
        dense_output=False
    )

    # compute conversion
    W_plot = sol.t
    F_CO = sol.y[0]
    X = (F_CO0 - F_CO) / F_CO0

    plt.plot(W_plot, X, label=f"{T_C} °C", linewidth=2)

# -------------------------------------------------------
# FINISH PLOT
# -------------------------------------------------------
plt.xlabel("Catalyst Weight W (kg)")
plt.ylabel("CO Conversion")
plt.title("Conversion vs Catalyst Weight at Different Temperatures")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```
import numpy as np
import scipy.integrate as spi
import matplotlib.pyplot as plt

def reactor_volume(T_C, P=40, x_final=0.65, rho_b=1500, F_A0=1515.22):
    """
    Calculate required catalyst weight and reactor volume.
    """

    # Temperature in Kelvin
    T_K = T_C + 273.15

    # Mole fractions
    y_CO  = 0.15
    y_H2  = 0.55
    y_CO2 = 0.05
    y_H2O = 0.20
    y_N2  = 0.02
    y_CH4 = 0.03

    y_total = y_CO + y_H2 + y_CO2 + y_H2O + y_N2 + y_CH4

    # Partial pressures (bar)
    p_A0 = P * (y_CO  / y_total)
    p_B0 = P * (y_H2O / y_total)

    # Constants
    k_eq = np.exp((4577.8 / T_K) - 4.33)
    k = 1.85e-5 * np.exp(-1855.5 / (8.314 * T_K))  # mol/g.min

    # Convert F_A0 from kmol/h → mol/min
    F_A0 = F_A0 * 1e3 / 60

    # Adsorption constants
    K_CO  = 0.1
    K_H2O = 0.2
    K_CO2 = 0.05
    K_H2  = 0.02

    # Rate expression for integration
    def rate_function(X):
        p_A = p_A0 * (1 - X)
        p_B = p_B0 * (1 - X)
        p_C = p_A0 * (y_CO2 / y_CO) * X
        p_D = p_A0 * (y_H2  / y_CO) * X

        numerator = p_A * p_B - (p_C * p_D / k_eq)
        denom = (1 + K_CO*p_A + K_H2O*p_B + K_CO2*p_C + K_H2*p_D)**2

        r_A = k * numerator / denom
        r_A = max(r_A, 1e-10)

        return F_A0 / r_A

    W, _ = spi.quad(rate_function, 0, x_final)
    V = W / rho_b / 1e3

    return W, V


def plot_X_vs_W(T_C=350, P=40, rho_b=1500, F_A0=1515.22):
    X_values = np.linspace(0, 0.65, 50)
    W_values = []

    for X in X_values:
        W, _ = reactor_volume(T_C=T_C, P=P, x_final=X,
                              rho_b=rho_b, F_A0=F_A0)
        W_values.append(W)

    plt.figure(figsize=(7,5))
    plt.plot(W_values, X_values, 'b-', linewidth=2)

    plt.xlabel("Catalyst Weight W (g)")
    plt.ylabel("Conversion X")
    plt.title("Conversion vs Catalyst Weight for PBR")
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# Run plot
plot_X_vs_W()
```

```
import numpy as np
import scipy.integrate as spi
import matplotlib.pyplot as plt

def reactor_volume(T_C, P=40, x_final=0.65, rho_b=1500, F_A0=1515.22):
    """
    Calculate required catalyst weight and reactor volume.

    Inputs:
        T_C   : Temperature in °C
        P     : Pressure in bar
        x_final : Final conversion
        rho_b : Bulk density of catalyst in kg/m^3
        F_A0  : Inlet molar flow rate of CO in kmol/h

    Returns:
        W_g   : Catalyst weight in g
        W_ton : Catalyst weight in metric tons
        V_m3  : Reactor volume in m^3
    """

    # Temperature in Kelvin
    T_K = T_C + 273.15

    # Mole fractions
    y_CO  = 0.15
    y_H2  = 0.55
    y_CO2 = 0.05
    y_H2O = 0.20
    y_N2  = 0.02
    y_CH4 = 0.03

    y_total = y_CO + y_H2 + y_CO2 + y_H2O + y_N2 + y_CH4

    # Partial pressures (bar)
    p_A0 = P * (y_CO  / y_total)   # CO
    p_B0 = P * (y_H2O / y_total)   # H2O

    # Constants
    k_eq = np.exp((4577.8 / T_K) - 4.33)

    # Rate constant (keep as you wrote it)
    k = 1.85e-5 * np.exp(-1855.5 / (8.314 * T_K))   # mol/g.min

    # Convert F_A0 from kmol/h to mol/min
    F_A0 = F_A0 * 1e3 / 60

    # Adsorption constants
    K_CO  = 0.1
    K_H2O = 0.2
    K_CO2 = 0.05
    K_H2  = 0.02

    def rate_function(X):
        p_A = p_A0 * (1 - X)
        p_B = p_B0 * (1 - X)
        p_C = p_A0 * (y_CO2 / y_CO) * X
        p_D = p_A0 * (y_H2  / y_CO) * X

        numerator = p_A * p_B - (p_C * p_D / k_eq)
        denom = (1 + K_CO*p_A + K_H2O*p_B + K_CO2*p_C + K_H2*p_D)**2

        r_A = k * numerator / denom   # mol/g.min

        # Avoid division by zero or negative values near equilibrium
        r_A = max(r_A, 1e-10)

        return F_A0 / r_A   # g

    # Catalyst weight in grams
    W_g, _ = spi.quad(rate_function, 0, x_final)

    # Convert to metric tons
    W_ton = W_g / 1e6

    # Reactor volume:
    # W_g -> kg = W_g / 1000
    # V = mass / density
    V_m3 = (W_g / 1000) / rho_b

    return W_g, W_ton, V_m3


def plot_X_vs_W(T_C=350, P=40, rho_b=1500, F_A0=1515.22):
    X_values = np.linspace(0.01, 0.65, 50)
    W_ton_values = []

    for X in X_values:
        W_g, W_ton, _ = reactor_volume(
            T_C=T_C, P=P, x_final=X, rho_b=rho_b, F_A0=F_A0
        )
        W_ton_values.append(W_ton)

    plt.figure(figsize=(7,5))
    plt.plot(W_ton_values, X_values, 'b-', linewidth=2)

    plt.xlabel("Catalyst Weight W (metric tons)")
    plt.ylabel("Conversion X")
    plt.title("Conversion vs Catalyst Weight for PBR")
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# Example run
W_g, W_ton, V_m3 = reactor_volume(T_C=350, x_final=0.65)
print(f"Catalyst weight = {W_g:.2f} g")
print(f"Catalyst weight = {W_ton:.6f} metric tons")
print(f"Reactor volume  = {V_m3:.6f} m^3")

plot_X_vs_W()
```

```
import numpy as np
import scipy.integrate as spi
import matplotlib.pyplot as plt

def reactor_weight_powerlaw(
    T_C=320,
    P_bar=30,
    X_final=0.75,
    k=1.0e-3
):
    """
    Packed-bed reactor sizing for HTS WGS using power-law kinetics.

    Basis from project file:
    CO + H2O <-> H2 + CO2

    Units:
    - T_C in degC
    - P_bar in bar (converted to atm for kinetics)
    - flowrates in kmol/h -> converted to mol/h
    - rate in mol/gcat.h
    - catalyst weight returned in g and ton
    """

    # -----------------------------
    # Given feed from file
    # -----------------------------
    F_CO0_kmh  = 1515.0
    F_H2O0_kmh = 2222.0
    F_H20_kmh  = 5555.0
    F_CO20_kmh = 505.0
    F_CH40_kmh = 303.0

    # Convert to mol/h
    F_CO0  = F_CO0_kmh  * 1e3
    F_H2O0 = F_H2O0_kmh * 1e3
    F_H20  = F_H20_kmh  * 1e3
    F_CO20 = F_CO20_kmh * 1e3
    F_CH40 = F_CH40_kmh * 1e3

    # Pressure in atm for kinetics
    P_atm = P_bar / 1.01325

    # Rate-law exponents from file
    n = 1.0
    m = 0.0
    l = -0.36
    p = -0.09

    def integrand(X):
        # Stoichiometric outlet flows
        F_CO  = F_CO0 * (1 - X)
        F_H2O = F_H2O0 - F_CO0 * X
        F_H2  = F_H20 + F_CO0 * X
        F_CO2 = F_CO20 + F_CO0 * X
        F_CH4 = F_CH40

        F_T = F_CO + F_H2O + F_H2 + F_CO2 + F_CH4

        # Mole fractions
        y_CO  = F_CO / F_T
        y_H2O = F_H2O / F_T
        y_H2  = F_H2 / F_T
        y_CO2 = F_CO2 / F_T

        # Partial pressures in atm
        P_CO  = y_CO * P_atm
        P_H2O = y_H2O * P_atm
        P_H2  = y_H2 * P_atm
        P_CO2 = y_CO2 * P_atm

        # Avoid numerical issues
        P_CO  = max(P_CO,  1e-12)
        P_H2O = max(P_H2O, 1e-12)
        P_H2  = max(P_H2,  1e-12)
        P_CO2 = max(P_CO2, 1e-12)

        # Power-law rate: mol/gcat.h
        r_CO = k * (P_CO**n) * (P_H2O**m) * (P_CO2**l) * (P_H2**p)

        r_CO = max(r_CO, 1e-12)

        # PBR design equation: dW/dX = F_CO0 / -r_CO
        return F_CO0 / r_CO

    W_g, _ = spi.quad(integrand, 0.0, X_final)
    W_ton = W_g / 1e6

    return W_g, W_ton


def plot_conversion_vs_weight(
    temps_C=[410, 420, 430, 440],
    P_bar=30,
    X_max=0.75,
    k_ref=1.0e-3
):
    """
    Plot X vs W using same power-law model.
    Here k is kept as a tunable constant unless you provide k(T).
    """

    plt.figure(figsize=(8, 6))

    for T_C in temps_C:
        X_vals = np.linspace(0.05, X_max, 60)
        W_vals = []

        # if no Arrhenius data is confirmed, keep same k
        k_T = k_ref

        for X in X_vals:
            _, W_ton = reactor_weight_powerlaw(
                T_C=T_C,
                P_bar=P_bar,
                X_final=X,
                k=k_T
            )
            W_vals.append(W_ton)

        plt.plot(W_vals, X_vals, linewidth=2, label=f"{T_C}°C")

    plt.xlabel("Catalyst Weight (metric tons)")
    plt.ylabel("CO Conversion X")
    plt.title("CO Conversion vs Catalyst Weight")
    plt.grid(True)
    plt.legend()
    plt.ticklabel_format(style='plain', axis='x')
    plt.tight_layout()
    plt.show()


# Example
W_g, W_ton = reactor_weight_powerlaw(T_C=320, P_bar=30, X_final=0.75, k=1.0e-3)
print(f"Catalyst weight = {W_g:.3e} g")
print(f"Catalyst weight = {W_ton:.2f} metric tons")

plot_conversion_vs_weight()
```

```
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Mole Fractions
y_CO = 0.15
y_H2O = 0.20
y_H2 = 0.55
y_CO2 = 0.05
y_CH4 = 0.03

# Operating conditions
F_A0 = 1515      # kmol/hr
T_C = 350        # °C
P = 40           # bar
rho_b = 1500     # kg/m3
X_final = 0.75   # target conversion


def reactor_volume(X_target):
    T_K = T_C + 273.15

    # convert kmol/hr → mol/min
    F_A0_mol = F_A0 * 1000 / 60

    # partial pressures
    p_CO0 = P * y_CO
    p_H2O0 = P * y_H2O

    # equilibrium constant
    Kp = np.exp((4577.8 / T_K) - 4.33)

    # kinetic constant
    k = 1.85e-5 * np.exp(-1855.5 / (8.314 * T_K))

    def rate_function(X):
        p_CO = p_CO0 * (1 - X)
        p_H2O = p_H2O0 * (1 - X)

        # stoichiometric products
        p_CO2 = p_CO0 * (y_CO2 / y_CO) * X
        p_H2  = p_CO0 * (y_H2  / y_CO) * X

        beta = (p_CO2 * p_H2) / (p_CO * p_H2O * Kp)

        r = k * p_CO * p_H2O * (1 - beta)
        r = max(r, 1e-10)

        return F_A0_mol / r

    # Integrate up to target conversion
    W_g, _ = quad(rate_function, 0, X_target)

    W_ton = W_g / 1_000_000        # g → ton
    V = W_g / rho_b / 1000         # g → m3

    return W_ton, V


def plot_X_W():
    X_values = np.linspace(0.01, X_final, 50)
    W_ton_values = []

    for X in X_values:
        W_ton, _ = reactor_volume(X)
        W_ton_values.append(W_ton)

    plt.figure(figsize=(7,5))
    plt.plot(W_ton_values, X_values, 'b-', linewidth=2)

    plt.xlabel("Catalyst Weight W (ton)")
    plt.ylabel("Conversion X")
    plt.title("Conversion vs Catalyst Weight (X–W Plot)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# Run main calculation
W, V = reactor_volume(X_final)
print("Catalyst weight (ton) =", format(W, '.6f'))
print("Reactor volume (m3) =", format(V, '.6f'))

# Produce plot
plot_X_W()
```

```
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Mole Fractions
y_CO = 0.15
y_H2O = 0.20
y_H2 = 0.55
y_CO2 = 0.05
y_CH4 = 0.03

# Operating conditions
F_A0 = 1515      # kmol/hr
X_final = 0.75   # target conversion
P = 40           # bar
rho_b = 1500     # kg/m3


def reactor_volume(T_C, X_target):
    T_K = T_C + 273.15
    F_A0_mol = F_A0 * 1000 / 60

    p_CO0 = P * y_CO
    p_H2O0 = P * y_H2O

    Kp = np.exp((4577.8 / T_K) - 4.33)
    k = 1.85e-5 * np.exp(-1855.5 / (8.314 * T_K))

    def rate_function(X):
        p_CO = p_CO0 * (1 - X)
        p_H2O = p_H2O0 * (1 - X)

        p_CO2 = p_CO0 * (y_CO2 / y_CO) * X
        p_H2  = p_CO0 * (y_H2  / y_CO) * X

        beta = (p_CO2 * p_H2) / (p_CO * p_H2O * Kp)
        r = k * p_CO * p_H2O * (1 - beta)
        r = max(r, 1e-10)

        return F_A0_mol / r

    W_g, _ = quad(rate_function, 0, X_target)
    return W_g / 1_000_000, W_g / rho_b / 1000


def plot_multi_temperature_subplots():
    temperatures = [310, 320, 380, 410]
    X_values = np.linspace(0.01, X_final, 50)

    fig, axs = plt.subplots(2, 2, figsize=(10, 8))
    axs = axs.flatten()

    for i, T_C in enumerate(temperatures):
        W_ton_values = []

        for X in X_values:
            W_ton, _ = reactor_volume(T_C, X)
            W_ton_values.append(W_ton)

        axs[i].plot(W_ton_values, X_values, linewidth=2)
        axs[i].set_title(f"{T_C}°C")
        axs[i].set_xlabel("Catalyst Weight W (ton)")
        axs[i].set_ylabel("Conversion X")
        axs[i].grid(True)
        axs[i].set_ylim(0.45, 0.78)  # Y‑axis zoom to make curves clearer

    fig.suptitle("Conversion vs Catalyst Weight for Different Temperatures", fontsize=14)
    plt.tight_layout()
    plt.show()


# Example single calculation
W, V = reactor_volume(350, X_final)
print("Catalyst weight (ton) =", format(W, '.6f'))
print("Reactor volume (m3) =", format(V, '.6f'))

# Plot separate subplots
plot_multi_temperature_subplots()
```
